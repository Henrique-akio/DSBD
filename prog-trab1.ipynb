```python
# Vamos obter informações sobre o rendimento de alunos de um curso em algumas disciplinas ao longo dos anos.
# O primeiro arquivo contido no Dataset Notas, historico-alg1_SIGA_ANONIMIZADO.csv, refere-se ao aproveitamento de estudantes na disciplina ALGORITMOS 1 entre os anos de 2011 e 2022.

# A primeira coluna ("matricula") é composta por números inteiros, onde cada número representa um indivíduo. Assim, repetições nessa coluna indicam que o estudante fez mais de uma vez a mesma matéria.

# Atenção: R-nota indica REPROVAÇÃO POR NOTA e R-freq REPROVAÇÃO POR FALTA. Se houver outro "status" para representar reprovação, este dever ser trocado para o rótulo adequado (R-nota ou R-freq). Frequências < 75 causam reprovação por falta; Médias abaixo de 50 causam reprovação por nota.
```

```python
# 1. Qual é a média de nota dos aprovados (no período total e por ano)?

df_filtered = df.reset_index()
df_filtered = df_filtered[
    (df_filtered['status'] == 'Aprovado') & 
    (df_filtered['tipo'] == 'TURMA') # outros são equivalencia e aproveitamento
    
]
df_filtered['total'] = 'total'

df_1_ano = df_filtered.groupby(['ano']).agg(        
    avg_nota = ('nota', 'mean'),
    median_nota = ('nota', 'median'),
    min_nota = ('nota', 'min'),
    max_nota = ('nota', 'max'),
    count = ('matricula', 'count')
    )

df_1_total = df_filtered.groupby(['total']).agg(        
    avg_nota = ('nota', 'mean'),
    median_nota = ('nota', 'median'),
    min_nota = ('nota', 'min'),
    max_nota = ('nota', 'max'),
    count = ('matricula', 'count')
    )

df_1_all = pd.concat([df_1_ano, df_1_total], axis=0)
df_1_all.round(1)

```

```python
# 2. Qual é a média de nota dos reprovados por nota (período total e ano)?

df_filtered = df.reset_index()
df_filtered = df_filtered[
    (df_filtered['status'].isin(['R-nota', 'R-freq'])) & 
    (df_filtered['tipo'] == 'TURMA')
    
]
df_filtered['total'] = 'total'

df_2_ano = df_filtered.groupby(['ano']).agg(        
    avg_nota = ('nota', 'mean'),
    median_nota = ('nota', 'median'),    
    min_nota = ('nota', 'min'),
    max_nota = ('nota', 'max'),
    count = ('matricula', 'count')
    )

df_2_total = df_filtered.groupby(['total']).agg(        
    avg_nota = ('nota', 'mean'),
    median_nota = ('nota', 'median'),
    min_nota = ('nota', 'min'),
    max_nota = ('nota', 'max'),
    count = ('matricula', 'count')
    )

df_2_all = pd.concat([df_2_ano, df_2_total], axis=0)
df_2_all.round(1)
```

```python
# 3. Qual é a frequência dos reprovados por nota (período total e por ano)?

df_filtered = df.reset_index()
df_filtered = df_filtered[
    (df_filtered['status'].isin(['R-nota'])) & 
    (df_filtered['tipo'] == 'TURMA')
    
]
df_filtered['total'] = 'total'

df_3_ano = df_filtered.groupby(['ano']).agg(        
    avg_freq = ('frequencia', 'mean'),
    median_freq = ('frequencia', 'median'),
    min_freq = ('frequencia', 'min'),
    max_freq = ('frequencia', 'max'),
    count = ('matricula', 'count')
    )

df_3_total = df_filtered.groupby(['total']).agg(        
    avg_freq = ('frequencia', 'mean'),
    median_freq = ('frequencia', 'median'),
    min_freq = ('frequencia', 'min'),
    max_freq = ('frequencia', 'max'),
    count = ('matricula', 'count')
    )

df_3_all = pd.concat([df_3_ano, df_3_total], axis=0)
df_3_all.round(1)
```

```python
# 4. Qual a porcentagem de evasões (total e anual)?
  
df_4_total = df
df_4_total['total'] = 'total'

df_4_ano = df.groupby('ano').agg(
    count_evasao=('situacaoDiscente', lambda x: (x == 'Evasão').sum()),
    count_all=('situacaoDiscente', 'size'), 
)

df_4_ano['ratio_evasao_all'] = df_4_ano['count_evasao'] / df_4_ano['count_all']

df_4_total = df.groupby('total').agg(
    count_evasao=('situacaoDiscente', lambda x: (x == 'Evasão').sum()),  
    count_all=('situacaoDiscente', 'size'),  
)    

df_4_total['ratio_evasao_all'] = df_4_total['count_evasao'] / df_4_total['count_all']

df_4_all = pd.concat([df_4_ano, df_4_total], axis=0)
df_4_all.round(2)
```

```python
# 5. Como os anos de pandemia impactaram no rendimento dos estudantes em relação aos anos anteriores, considerando o rendimento dos aprovados, 
# a taxa de cancelamento e as reprovações? Considere como anos de pandemia os anos de 2020 e 2021.

# 1 - plotar a distribuição de cada um deles

df_filtered = df.reset_index()

df_filtered['grupo_ano'] = df_filtered.apply(
    lambda x: '2018-2019' if x['ano'] in [2018, 2019] 
    else ('2020-2021' if x['ano'] in [2020, 2021] 
    else ('2022-1' if x['ano'] == 2022 and x['periodo'] == '1' 
    else ('2022-2' if x['ano'] == 2022 and x['periodo'] == '2' 
    else 'Outros'))),
    axis=1)

df_filtered = df_filtered[
    (df_filtered['status'] != 'Matriculado') & # parece ser casos que a disciplina ainda está em aberto
    (df_filtered['tipo'] == 'TURMA') & # outros são equivalencia e aproveitamento
    (df_filtered['grupo_ano'].isin(['2018-2019', '2020-2021'])) # outros são equivalencia e aproveitamento
]


df_5 = df_filtered.groupby(['grupo_ano']).agg(        
    pct_aprovado=('status', lambda x: (x == 'Aprovado').sum() / len(x) * 100),
    pct_reprovado_nota=('status', lambda x: (x == 'R-freq').sum() / len(x) * 100),
    pct_reprovado_freq=('status', lambda x: (x == 'R-nota').sum() / len(x) * 100),
    pct_cancelado=('status', lambda x: (x == 'Cancelado').sum() / len(x) * 100),
    pct_reprovado=('status', lambda x: (x == 'Reprovado').sum() / len(x) * 100),
    pct_matriculado=('status', lambda x: (x == 'Matriculado').sum() / len(x) * 100),     
    avg_nota = ('nota', 'mean'),       
    avg_nota_aprovados=('nota', lambda x: x[df_filtered.loc[x.index, 'status'] == 'Aprovado'].mean()),
    avg_nota_reprovados_nota =('nota', lambda x: x[df_filtered.loc[x.index, 'status'].isin(['R-nota'])].mean()),     
    avg_nota_reprovados_freq =('nota', lambda x: x[df_filtered.loc[x.index, 'status'].isin(['R-freq'])].mean()),     
    count_aprovado=('status', lambda x: (x == 'Aprovado').sum()),
    count_reprovado_nota=('status', lambda x: (x == 'R-freq').sum()),
    count_reprovado_freq=('status', lambda x: (x == 'R-nota').sum()),
    count_cancelado=('status', lambda x: (x == 'Cancelado').sum()),
    count_reprovado=('status', lambda x: (x == 'Reprovado').sum()),
    count_matriculado=('status', lambda x: (x == 'Matriculado').sum()),    
    count=('matricula', 'count'),
    )

dif_percentuais = df_5.loc['2020-2021'] - df_5.loc['2018-2019']
dif_df = pd.DataFrame([dif_percentuais], index=['Dif_Percentual'])
df_5_final = pd.concat([df_5, dif_df])
df_5_final.round(1)
```

```python
# 6. Compare a volta às aulas híbrida (2022 período 1) com os anos de pandemia e os anos anteriores.

df_filtered = df.reset_index()

df_filtered['grupo_ano'] = df_filtered.apply(
    lambda x: '2018-2019' if x['ano'] in [2018, 2019] 
    else ('2020-2021' if x['ano'] in [2020, 2021] 
    else ('2022-1' if x['ano'] == 2022 and x['periodo'] == '1' 
    else ('2022-2' if x['ano'] == 2022 and x['periodo'] == '2' 
    else 'Outros'))),
    axis=1)

df_filtered = df_filtered[
    (df_filtered['status'] != 'Matriculado') & # parece ser casos que a disciplina ainda está em aberto    
    (df_filtered['tipo'] == 'TURMA') & # outros são equivalencia e aproveitamento
    (df_filtered['grupo_ano'].isin(['2018-2019', '2020-2021', '2022-1', '2022-2'])) # outros são equivalencia e aproveitamento
]


df_5 = df_filtered.groupby(['grupo_ano']).agg(        
    pct_aprovado=('status', lambda x: (x == 'Aprovado').sum() / len(x) * 100),
    pct_reprovado_nota=('status', lambda x: (x == 'R-freq').sum() / len(x) * 100),
    pct_reprovado_freq=('status', lambda x: (x == 'R-nota').sum() / len(x) * 100),
    pct_cancelado=('status', lambda x: (x == 'Cancelado').sum() / len(x) * 100),
    pct_reprovado=('status', lambda x: (x == 'Reprovado').sum() / len(x) * 100),
    pct_matriculado=('status', lambda x: (x == 'Matriculado').sum() / len(x) * 100),     
    avg_nota = ('nota', 'mean'),       
    avg_nota_aprovados=('nota', lambda x: x[df_filtered.loc[x.index, 'status'] == 'Aprovado'].mean()),
    avg_nota_reprovados_nota =('nota', lambda x: x[df_filtered.loc[x.index, 'status'].isin(['R-nota'])].mean()),     
    avg_nota_reprovados_freq =('nota', lambda x: x[df_filtered.loc[x.index, 'status'].isin(['R-freq'])].mean()),     
    count_aprovado=('status', lambda x: (x == 'Aprovado').sum()),
    count_reprovado_nota=('status', lambda x: (x == 'R-freq').sum()),
    count_reprovado_freq=('status', lambda x: (x == 'R-nota').sum()),
    count_cancelado=('status', lambda x: (x == 'Cancelado').sum()),
    count_reprovado=('status', lambda x: (x == 'Reprovado').sum()),
    count_matriculado=('status', lambda x: (x == 'Matriculado').sum()),    
    count=('matricula', 'count'),
    )

dif_2022_1__2018_2019 = df_5.loc['2022-1'] - df_5.loc['2018-2019']
dif_2022_1__pandemia = df_5.loc['2022-1'] - df_5.loc['2020-2021']
dif_df_normal = pd.DataFrame([dif_2022_1__2018_2019], index=['dif_2022_1__2018_2019'])
dif_df_pandemia = pd.DataFrame([dif_2022_1__pandemia], index=['dif_2022_1__pandemia'])
df_5_final = pd.concat([df_5, dif_df_normal, dif_df_pandemia])
df_5_final.round(1)
```
